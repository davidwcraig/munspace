```python
import numpy as np
import matplotlib.pyplot as plt
from munspace import hvc, midhues
```

```python
def gau(x, loc=0.0, scale=1.0):
    return np.exp( -((x-loc)/scale)**2 )
```

```python
mr = hvc('5RP',4,'high')
mg = hvc('5G',4, 'high')
```

```python
X = np.linspace(-5,5,100)
```

```python
Y = gau(X)
Y1 = 0.5*gau(X,scale=2)
```

```python
plt.plot(X,Y,color=mr)
plt.plot(X,Y1,color=mg)
```

```python
plt.scatter(X,Y, c=[hvc(hue,4,'middle') for hue in midhues]*10)
```

```python
from cycler import cycler
```

**See:** <https://matplotlib.org/3.1.1/tutorials/intermediate/color_cycle.html>

```python
muncolors = [ hvc(hue, 4, 'high') for hue in midhues]
```

```python
muncolors
```

```python
default_cycler = (cycler(color=muncolors))
```

```python
plt.rc('lines', linewidth=4)
plt.rc('axes', prop_cycle=default_cycler)
```

```python
x = np.linspace(0, 2 * np.pi, 50)
offsets = np.linspace(0, 2 * np.pi, 10, endpoint=False)
yy = np.transpose([np.sin(x + phi) for phi in offsets])
```

```python
plt.plot(yy)
plt.savefig("munsell-color-demo.png")
```
